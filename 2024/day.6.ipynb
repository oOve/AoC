```python
import pandas as pd
import numpy as np
import re

from IPython.display import display_html, clear_output
import time
```

```python
import aocd
import os
from functools import reduce

session=open("secret.session").read()
os.environ["AOC_SESSION"] = session
data = aocd.get_data(day=6,year=2024)
```

```python
data_map = np.asarray([list(d) for d in data.splitlines()])
```

```python
def show(a):
    clear_output(wait=True)
    display_html('<pre style="font-size: 16px;line-height:60%;overflow: auto;">' + '<br>'.join([''.join(x) for x in a.tolist()]) + '</pre>', raw=True)
    time.sleep(0.005)

'''
╔ ╗
╚ ╝
'''
glyphs = {
    (-1, 0) : '╔',
    ( 0, 1):  '╗',
    ( 1, 0):  '╝',
    ( 0,-1):  '╚'
}


# Part 1
cmap = data_map.copy()
cmap[cmap=='.'] = ' '
heading  = np.asarray([-1,0])
position = np.where(data_map=='^')
position = np.asarray([position[0][0],position[1][0]])
size = data_map.shape

while True:
    
    show(cmap)
    if (position<0).any() or (position>size).any():
        break

    if (cmap[*(position+heading)] == '#'):
        cmap[*position] = glyphs[tuple(heading)]
        heading[0],heading[1]=heading[1],heading[0]*-1
    else:
        position=position+heading
        cmap[*position] = '║' if heading[1]==0 else '═'

print()
 
np.sum(cmap=='X')
```
