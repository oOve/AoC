```python
import pandas as pd
import numpy as np
import re

from IPython.display import display_html, clear_output
import time
```

```python
import aocd
import os
from functools import reduce

session=open("secret.session").read()
os.environ["AOC_SESSION"] = session
data = aocd.get_data(day=6,year=2024)
```

```python
data_map = np.asarray([list(d) for d in data.splitlines()])
```

```python
# Part 1
cmap = data_map.copy()
heading  = np.asarray([-1,0])
position = np.where(data_map=='^')
position = np.asarray([position[0][0],position[1][0]])
size = data_map.shape

while True:
    cmap[*position] = 'X'
    new_pos = position+heading
    if (new_pos<0).any() or (new_pos>size).any():
        break    

    if (cmap[*new_pos] == '#'):
        heading[0],heading[1]=heading[1],heading[0]*-1
    else:
        position=position+heading

np.sum(cmap=='X')
```

```python
# Part 2
left  = 1
right = 2
up    = 4
down  = 8
new_obstruction = 16

lefth = ( 0, 1)
righth= ( 0,-1)
uph   = (-1, 0)
downh = ( 1, 0)

heading2bit = {
    lefth:left,
    righth:right,
    uph:up,
    downh:down
}

def show2(a):
    clear_output(wait=True)
    a2= np.logical_not(a).astype('<U1')
    a2[a2=='T']=' '
    a2[a2!=' ']='X'
    display_html('<pre style="font-size: 10px;font-stretch:200%;line-height:90%;overflow: auto;white-space: pre;overflow: hidden;">' + '<br>'.join([''.join(x) for x in a2.tolist()]) + '</pre>', raw=True)


def is_cyclic(bmap, barriers, pos, head):
    while True:
        hbit=heading2bit[tuple(head)]
        if bmap[*pos] & hbit:
            return True
        bmap[*pos] |= hbit

        npos = pos+head
        if (npos<0).any() or (npos>=size).any():
            return False
        if barriers[*npos]:
            head[0],head[1]=head[1],head[0]*-1
        else:
            pos=npos
        


heading  = np.asarray(uph)
start    = np.where(data_map=='^')
start    = np.asarray([start[0][0],start[1][0]])
position = start

size = data_map.shape

new_barrier_options = set()

matrix   = np.zeros(size, dtype=int)
barriers = data_map == '#'

while True:
    new_pos = position+heading

    if (new_pos<0).any() or (new_pos>=size).any():
        break
    
    if barriers[*new_pos]:
        heading[0],heading[1]=heading[1],heading[0]*-1
    else:

        # Should we introduce a barrier?
        if matrix[*new_pos] == 0:
            bc = barriers.copy()
            bc[*new_pos] = True
            mc = matrix.copy()
            cyclic = is_cyclic(mc, bc, position.copy(), heading.copy())
            if cyclic:
                new_barrier_options.add(tuple(new_pos))
                show2(mc)
        hbit = heading2bit[tuple(heading)]
        matrix[*position] |= hbit
        position=new_pos
    
    show2(matrix)

a = np.zeros(size, dtype='<U1')
a[:]=' '
for p in new_barrier_options:
    a[*p]='O'

show(a)
print( len(new_barrier_options) )
```
